```python
import numpy as np
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def link_tokens(docs):
    corpus = list()
    for doc in docs:
        corpus.append(' '.join(doc))
    return corpus
```

```python
corpus = np.load("/Users/zhang/MscProject_tweak2vec/corpus/stack_corpus_withunk.npy")
corpus = corpus.tolist()
corpus = link_tokens(corpus)
```

```python
vectorizer = CountVectorizer()
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(corpus)) 
```

```python
weightlist = tfidf.toarray()
wordlist = vectorizer.get_feature_names() # all the words in bag-of-words model
```

```python
for i in range(100):  
    print("-------这里输出第",i,"类文本的词语tf-idf权重------")
    for j in range(len(wordlist)):  
        print(wordlist[j],weightlist[i][j])
```
