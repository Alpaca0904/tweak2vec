```python
import numpy as np
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm
from sklearn.multiclass import OneVsRestClassifier
```

```python
def link_tokens(docs):
    corpus = list()
    for doc in docs:
        corpus.append(' '.join(doc))
    return corpus
```

```python
corpus = np.load("/Users/zhang/MscProject_tweak2vec/corpus/stack_corpus_withunk200.npy")
corpus = corpus.tolist()
corpus = link_tokens(corpus)
```

```python
vectorizer = CountVectorizer()
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(corpus)) 
```

```python
weightlist = tfidf.toarray()
wordlist = vectorizer.get_feature_names() # all the words in bag-of-words model
```

```python
stack_corpus = weightlist
```

```python
for i in range(100):  
    print("-------这里输出第",i,"类文本的词语tf-idf权重------")
    for j in range(len(wordlist)):  
        print(wordlist[j],weightlist[i][j])
```

```python
traindatasize_1 = 1000
traindatasize_2 = 500

stack_train_tfidf_x = np.concatenate((stack_corpus[0:0+traindatasize_1],
                             stack_corpus[50000:50000+traindatasize_1],
                             stack_corpus[100000:100000+traindatasize_2],
                             stack_corpus[150000:150000+traindatasize_2],
                             stack_corpus[175000:175000+traindatasize_2]), axis=0)

stack_train_y = np.concatenate((np.ones(traindatasize_1)*0,
                                np.ones(traindatasize_1)*1,
                                np.ones(traindatasize_2)*2,
                                np.ones(traindatasize_2)*3,
                                np.ones(traindatasize_2)*4), axis=0)
```

```python
valdatasize_1 = 400
valdatasize_2 = 200

stack_val_tfidf_x = np.concatenate((stack_corpus[40000:40000+valdatasize_1],
                             stack_corpus[90000:90000+valdatasize_1],
                             stack_corpus[140000:140000+valdatasize_2],
                             stack_corpus[170000:170000+valdatasize_2],
                             stack_corpus[195000:195000+valdatasize_2]), axis=0)

stack_val_y = np.concatenate((np.ones(valdatasize_1)*0,
                              np.ones(valdatasize_1)*1,
                              np.ones(valdatasize_2)*2,
                              np.ones(valdatasize_2)*3,
                              np.ones(valdatasize_2)*4), axis=0)
```

```python
model = OneVsRestClassifier(svm.SVC(kernel='linear'))
clf_tfidf = model.fit(stack_train_tfidf_x,stack_train_y)
```

```python
clf_tfidf.score(stack_train_tfidf_x, stack_train_y)
```

```python
clf_tfidf.score(stack_val_tfidf_x, stack_train_y)
```
