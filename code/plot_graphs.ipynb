```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_data(filename):
    f = open(filename)
    train_los = []
    train_acc = []
    val_los = []
    val_acc = []
    i = 1
    for line in f:    
        if i%2==0:
            data = line.split()
            train_los.append(float(data[4]))
            train_acc.append(float(data[7]))
            val_los.append(float(data[10]))
            val_acc.append(float(data[13]))
        i = i+1
    return train_los, train_acc, val_los, val_acc
```

```python
results = {}
# results['google pretrain 300d'] = get_data('/Users/zhang/MscProject_tweak2vec/data/1_pretrain.txt')
results['google pretrain'] = get_data('/Users/zhang/MscProject_tweak2vec/data/1_pretrain_50d.txt')
results['quora corpus'] = get_data('/Users/zhang/MscProject_tweak2vec/data/2_quora_50d.txt')
# results['100 pivots reg'] = get_data('/Users/zhang/MscProject_tweak2vec/data/6_100pivots_reg_50d.txt')
# results['vec concatenate'] = get_data('/Users/zhang/MscProject_tweak2vec/data/3_concat.txt')
# results['vec average'] = get_data('/Users/zhang/MscProject_tweak2vec/data/4_avg.txt')
# results['google retrain'] = get_data('/Users/zhang/MscProject_tweak2vec/data/5_retrain.txt')
# results['100 pivots freeze'] = get_data('/Users/zhang/MscProject_tweak2vec/data/6_100pivots.txt')
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = plt.subplot(111)
epoch_num = 50
for k in results.keys():
    #ax1.plot(np.arange(epoch_num),results[k][1], linestyle=':', label=k)
    ax1.plot(np.arange(epoch_num),results[k][3], linestyle='-', label=k)
ax1.legend(loc=0)
ax1.set_xlabel('Epoch number')
ax1.set_ylabel('Validation Accuarcy')
plt.show()
```
